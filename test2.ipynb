```python
import pandas as pd
```

```python
nifty_data = pd.read_csv('untitled spreadsheet - Sheet1.csv')
portfolio_data  = pd.read_csv('untitled spreadsheet - Sheet2.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
nifty_data_selected = nifty_data[['Daily Return','Weekly Return','Monthly Return']]
portfolio_data = portfolio_data.rename(columns={'Unnamed: 6' : 'Daily Return nifty' ,'Unnamed: 7' : 'Weekly Return nifty' , 'Unnamed: 8' : 'Monthly Return nifty' })
```

```python
portfolio_data_selected = portfolio_data[['Daily Return nifty','Weekly Return nifty','Monthly Return nifty']]

combined_df = pd.concat([nifty_data_selected, portfolio_data_selected], axis=1)
```

```python

correlation_matrix = combined_df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Plot Between Selected Columns")
plt.show()
```
